### Task 3: Visualizaton of Employee Distributon

###  Create bar charts to visualize the distributon of employees across departments and job 
### roles.

```python
plt.figure(figsize=(10,5))
sb.countplot(x='Department', data=df,hue=df.Department, palette='pastel')
plt.title('Employee Distribution Across Department')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
sb.countplot(x='JobRole', data=df,hue=df.JobRole,palette='pastel')
plt.xticks(rotation=90)
plt.title('Employee Distribution Across Job Roles')
plt.xlabel('Job Role') 
plt.ylabel('Count')
plt.show()
```

###  Use histograms to understand the distributon of numeric variables such as age and salary

```python
plt.figure(figsize=(10,5))
sb.histplot(x='Age',data=df,color='brown')
plt.xticks(rotation=90)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sb.histplot(x='MonthlyIncome',data=df,color='brown')
plt.xticks(rotation=90)
plt.title('Monthly Income Distribution')
plt.xlabel('Monthly Income')
plt.ylabel('Frequency')
plt.show()
```

# Task 4: Correlation Analysis

### Compute the correlaton matrix using NumPy.

```python
a=df.corr(numeric_only=True)
a
```

### Visualize correlatons using Seaborn’s heatmap functon.

```python
plt.figure(figsize=(20,10))
sb.heatmap(a,annot=True ,fmt=".2f", cmap="Pastel1",linewidths=0.2,annot_kws={"size": 8})
plt.title('Correlation Matrix') 
plt.show()
```

###  Compare satisfacton levels, salaries, and performance ratings between retained and attrited 
### employees using box plots. 


```python
plt.figure(figsize=(16,5))

plt.subplot(1,3,1)
sb.boxplot(data=df, x='Attrition',y='MonthlyIncome',hue=df.Attrition ,palette='pastel')

plt.subplot(1,3,2)
sb.boxplot(data=df, x='Attrition',y='PerformanceRating',hue=df.Attrition,palette='pastel')

plt.subplot(1,3,3)
sb.boxplot(data=df, x='Attrition',y='JobSatisfaction',hue=df.Attrition,palette='pastel')

plt.tight_layout()
plt.show()
```

### Identify trends in attriton rates across different demographics

```python
plt.figure(figsize=(16,5))

plt.subplot(2,2,1)
sb.countplot(x=df.Age,hue=df.Attrition,palette='pastel')
plt.xticks(rotation=90)

plt.subplot(2,2,2)
sb.countplot(x=df.Gender,hue=df.Attrition,palette='pastel')
plt.xticks(rotation=90)

plt.subplot(2,2,3)
sb.countplot(x=df.EducationField,hue=df.Attrition,palette='pastel')
plt.xticks(rotation=90)

plt.subplot(2,2,4)
sb.countplot(x=df.Department,hue=df.Attrition,palette='pastel')
plt.xticks(rotation=90)

plt.show()
```
